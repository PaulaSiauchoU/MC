```python
fig=figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(range(10))
print "getp(fig.patch)"
getp(fig.patch) # This prints the Artist's properties
print "getp(ax.patch)"
getp(ax.patch)
show(fig)
```
