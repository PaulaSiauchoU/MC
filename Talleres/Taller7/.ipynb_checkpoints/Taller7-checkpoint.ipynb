```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
from scipy.integrate import odeint
from scipy.integrate import ode
```

```python
#constantes
a = 0.1
b = 0.1
c = 14

def ros(Y,t):
    return np.array([-Y[1]-Y[2],Y[0]+(a*Y[1]),b+Y[2]*(Y[0]-c)])
```

```python
tmax=600
tes=np.linspace(0,tmax,10000)
rossol,others=odeint(ros,[1.,1.,1.],tes,full_output=True,printmessg=True)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure(figsize=(20,20))
ax = fig.gca(projection='3d')
plt.title("odeint")
ax.plot(rossol[:,0],rossol[:,1],rossol[:,2],label="error")
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.show()

plt.figure(figsize=(20,20))
plt.plot(rossol[:,0],rossol[:,1],color='blueviolet')
```
