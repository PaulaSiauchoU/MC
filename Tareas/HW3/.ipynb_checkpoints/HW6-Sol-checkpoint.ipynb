## Ejercicio 1

```python
higgs = genfromtxt("/Users/juan/Dropbox/UAndes/HerramientasComputacionalesPrivate/Homeworks/HW6/\
higgs.csv",delimiter=",")
dataenergies = higgs[:,0]
dataevents = higgs[:,1]
def nohig(x):
    return 27890 - 526.3 * x +  3.410* x** 2 - 0.007502* x **3
def yeshig(x):
    return nohig(x) + 116 * exp(-(x-125)**2/4.2)
energies = linspace(105,153,100)
nohiggsev = nohig(energies)
yeshiggsev = yeshig(energies)
```

```python
myfig, ax = subplots(figsize=(8, 8))
rc("font",size=15)
errorbar(dataenergies, dataevents,yerr=35,marker='o',
             linestyle='None',color="black",label="Data")
plot(energies,nohiggsev,'r--',label="$Ev$",linewidth=2)
plot(energies,yeshiggsev,'b',label="$Ev^*$",linewidth=2)
title("CMS $\sqrt{s}=7\,\mathrm{TeV}$, $L=5.1\,\mathrm{fb}^{-1}$ $\sqrt{s}=8\,\mathrm{TeV}$, $L = 5.3\,\mathrm{fb}^{-1}$")
minorticks_on()
tick_params('both', length=10, width=1, which='major')
tick_params('both', length=5, width=1, which='minor')
xlabel("$m_{\gamma\gamma}$(GeV)")
ylabel("S/(S+B) Weighted Events / $1.5$ GeV")
ylim([0,1900])
xlim([105,153])
legend(loc=1,frameon=False)
savefig("/Users/juan/Dropbox/UAndes/HerramientasComputacionalesPrivate/Homeworks/HW6/CMS-Higgs.pdf")
show()
```

# Ejercicio 2

```python
figure(figsize=(15,15))
# Fermat's Spiral
subplot(2,2,1)
angulos = linspace(0,100,1000)
radios = sqrt(angulos)
plot(radios*cos(angulos),radios*sin(angulos),"b",lw=2)
plot(-radios*cos(angulos),-radios*sin(angulos),"g",lw=2)
axis("equal")
title("Fermat's Spiral")
# Equiangular Spiral
subplot(2,2,2)
angulos = linspace(0,10*pi,1000)
b = 7 * pi / 16
radios = exp(angulos/tan(b))
plot(radios*cos(angulos),radios*sin(angulos),"b:",lw=4)
axis("equal")
title("Equiangular Spiral")
# Spiral of Archimedes
subplot(2,2,3)
angulos = linspace(0,10*pi,1000)
radios = angulos
plot(radios*cos(angulos),radios*sin(angulos),"r-.",lw=5)
axis("equal")
title("Spiral of Archimedes")
# Hyperbolic Spiral
subplot(2,2,4)
angulos = linspace(1,10*pi,1000)
radios = 1/angulos
plot(radios*cos(angulos),radios*sin(angulos),"m^",ms=5)
axis("equal")
title("Hyperbolic Spiral")
rc("font",size=20)
savefig("/Users/juan/Dropbox/UAndes/HerramientasComputacionalesPrivate/Homeworks/HW6/espirales.pdf")
show()
```
