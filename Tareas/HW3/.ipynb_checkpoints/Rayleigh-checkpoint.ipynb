##Introducción 

La distribución de probabilidad de Rayleigh, llamada así por Lord Rayleigh, es la distribución de la magnitud de un vector bidimensional cuyas coordinadas son independientes, distribuidas idénticamente,variables normalizadas con promedio cero. Esta distribución es importante en aplicaciones donde sus ajustes tienen variables normales relevantes.(Alabama University,sf)
En este caso se quiere estudiar la distribución de un gas de discos suaves que interactúan de acuerdo al potencial de Lennard Jones truncado. Éste último se caracteriza por modelar matemáticamente la aproximación de la interacción de dos átomos neutros o moléculas (Lennard-Jones). Para lograr el objetivo se hicieron histogramas con el comportamiento de la velocidad en X y Y así como la rapidez de las partículas para saber si ésta tendía a la distribución de Rayleigh.
