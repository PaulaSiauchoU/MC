```python
from scipy.fftpack import ifft, fft, fftfreq, fftshift
```

```python
#Se crea el array de ceros. 
placa=numpy.zeros((1024,1024))
#En el rango del for se tiene que tener en cuenta que se empieza a contar desde 0.Los límites se obtienen dividiendo 
#1024 entre 2 y después restándole o sumándole 
for i in range (496,528):
    for j in range (496,528):
        placa[i][j]=1
```

```python
imshow(placa)
```

```python
#Transformada de Fourier del cuadrado
TransPlaca=fft2(placa)
#Cambio de coordenadas 
SPlaca=fftshift(TransPlaca)
#Módulo 
ap=abs(SPlaca)
mod=sqrt(ap)
#logaritmo 
lp=log(mod)
#Creación de un array vacío para llenarlo después con los valores correspondientes de lp
final=numpy.zeros((200,200))
for i in range(412,612):
    for j in range(412,612):
        final[i-412][j-412]=lp[i][j]
print final
imshow(final)
```
